```python
#Given proababilities

p_disease=0.01
p_positive_given_disease=0.98
p_negative_given_nodisease=0.90
```

```python
#Calculation of compliment probabilities

p_nodisease= 1-p_disease
p_positive_given_nodisease=1-p_negative_given_nodisease
```

```python
#Bayes Theorem
p_disease_given_positive=(p_positive_given_disease * p_disease)/((p_positive_given_disease * p_disease)+(p_positive_given_nodisease * p_nodisease))
print("The probability of having the disease given a positive test result: ",p_disease_given_positive)
```

```python
p_disease_given_positive=(p_positive_given_disease * p_disease)/(p_positive_given_disease * p_disease)+(p_positive_given_nodisease * p_disease)
print(p_disease_given_positive)
```

```python
percentage=p_disease_given_positive*100
print(percentage)
```
